```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars=pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],height=df_used_cars["Cars Listings"],color="blue")
plt.xticks(rotation=45)
plt.title("Car listings by brand")
plt.show()
plt.savefig("usedcars.png")
```

```python
df_fuel_engine_types=pd.read_csv("stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
plt.figure(figsize=(9,6))
labels=["Diesel","Pertol","Gas"]
plt.stackplot(df_fuel_engine_types["Year"],
             df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"])
plt.xticks(df_fuel_engine_types["Year"],rotation=90)
plt.legend(labels=labels,loc="upper left")
plt.title("popularity of fuel types",fontsize=14)
plt.show()
```

```python
df_spx_fse_00_10=pd.read_csv("line_chart_data.csv")
```

```python
df_spx_fse_00_10
```

```python
df_spx_fse_00_10["new_date"]=pd.to_datetime(df_spx_fse_00_10["Date"])
```

```python
df_spx_fse_00_10["new_date"]
```

```python
labels=["S&P 500", "FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(df_spx_fse_00_10["new_date"],df_spx_fse_00_10["GSPC500"])
plt.plot(df_spx_fse_00_10["new_date"],df_spx_fse_00_10["FTSE100"])
plt.title("S&P vs FTSE Returns",fontsize=14,fontweight="bold")
plt.ylabel("returns")
plt.xlabel("Date")
plt.legend(labels=labels,fontsize="large")
plt.show()
```

```python
df_real_estate=pd.read_csv("histogram_data.csv")
```

```python
df_real_estate
```

```python
sns.set_style("white")
plt.figure(figsize=(8,6))
plt.hist(df_real_estate["Price"],bins=8,color="green")
plt.title("Distribution of Prices",fontsize=14,weight="bold")
plt.xlabel("price in (000' $)")
plt.ylabel("number of properties")
sns.despine()
plt.show()
```

```python
df_real_estate=pd.read_csv("Scatter_data.csv")
```

```python
df_real_estate
```

```python
plt.figure(figsize=(12,8))
scatter=plt.scatter(df_real_estate['Area (ft.)'],
           df_real_estate['Price'],alpha=0.6,c=df_real_estate['Building Type']
           )
plt.legend(*scatter.legend_elements(),loc="upper left",
          title="Building Type")
plt.title("Relationship between Area and Price of California Real Estate",fontsize=14,weight="bold")
plt.xlabel("Area (sq.ft)",weight="bold")
plt.ylabel("Area (sq.ft)",weight="bold")
plt.show()
```
